```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
y = np.array([0,1,1,0])

print("Input:\n",X)
print("Output:\n",y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='coolwarm')
plt.title("XOR Dataset")
plt.xlabel("Features 1")
plt.ylabel("Features 2")
plt.show()
```

```python
np.random.seed(42)

input_size = 2
hidden_size = 3
output_size = 1

W1 = np.random.randn(input_size,hidden_size)
b1 = np.zeros((1,hidden_size))

W2 = np.random.randn(hidden_size,output_size)
b2 = np.zeros((1,output_size))

print("W1:\n",W1)
print("W2:\n",W2)
```

```python
plt.imshow(W1,aspect='auto')
plt.colorbar()
plt.title("Weights: Input -> Hidden (W1) ")
plt.xlabel("Hidden Nodes")
plt.ylabel("Input Nodes")
plt.show()

plt.imshow(W2,aspect='auto')
plt.colorbar()
plt.title("Weights: Hidden -> Output (W2)")
plt.xlabel("Output Nodes")
plt.ylabel("Hidden Nodes")
plt.show()
```

```python
def sigmoid(x):
  return 1/ (1+np.exp(-x))
```

```python
def forward(X):
  z1 = np.dot(X,W1) + b1
  a1 = sigmoid(z1)

  z2 = np.dot(a1,W2) + b2
  a2 = sigmoid(z2)

  return z1,a1,z2,a2
```

```python
z1, a1 , z2 , output = forward(X)

print("Hidden Layer Output:\n",a1)
print("Final Output:\n",output)
```

```python
plt.imshow(a1,aspect='auto')
plt.colorbar()
plt.title("Hidden Layer Activations")
plt.xlabel("Hidden Nodes")
plt.ylabel("Input Samples")
plt.show()
```

```python
for i in range(len(X)):
  print(f"\n Input: {X[i]}")
  print(f"Hidden Activations: {a1[i]}")
  print(f"Output: {output[i]}")
```

```python
plt.bar(range(len(y)),y.flatten(),alpha=0.6,label='Actual')
plt.bar(range(len(output)),output.flatten(),alpha=0.6,label='Predicted')
plt.legend()
plt.title("Actual vs Predicted")
plt.show()
```

```python
loss = np.mean((y - output) ** 2)
print("Loss:",loss)
```

```python
learning_rate = 0.1
losses = []

for epoch in range(10000):
  z1, a1, z2, output = forward(X)

  loss = np.mean((y - output) ** 2 )
  losses.append(loss)
  error = y - output
  d_output = error * output * (1 - output)
  d_hidden = np.dot(d_output, W2.T) * a1 * (1 - a1)

  W2 += a1.T.dot(d_output) * learning_rate
  W1 += X.T.dot(d_hidden) * learning_rate
```

```python
plt.plot(losses)
plt.title("Loss Decreasing Over Time")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

```python
_, _, _, predicitions = forward(X)
print("Prediction",predicitions)
```

```python
plt.bar(range(len(y)),y.flatten(),alpha=0.6,label='Actual')
plt.bar(range(len(predicitions)),predicitions.flatten(),alpha=0.6,label='Predicted')
plt.legend()
plt.title("After Trainging: Prediction vs Actual")
plt.show()
```

```python
xx , yy = np.meshgrid(np.linspace(-0.5,1.5,100),np.linspace(-0.5,1.5,100))

grid = np.c_[xx.ravel(), yy.ravel()]
_, _, _, preds = forward(grid)

z = preds.reshape(xx.shape)

plt.contourf(xx,yy,z,cmap='coolwarm',alpha=0.5)
plt.scatter(X[:,0],X[:,1],c=y.flatten(),edgecolors='k',cmap='coolwarm')
plt.title("Decison Boundary")
plt.show()
```
