```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n = 200

study_hours = np.random.randint(1,11,n)
sleep_hours = np.random.randint(0,11,n)

score = study_hours * 0.7 + sleep_hours * 0.3

result = (score > 6).astype(int)

data = pd.DataFrame({
    'Study Hours': study_hours,
    'Sleep Hours': sleep_hours,
    'Result': result
})

data.head()
```

```python
plt.scatter(data['Study Hours'], data['Sleep Hours'],c=data['Result'])
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.title('Pass 1 vs Fail 0')
plt.show()
```

```python
X = data[['Study Hours', 'Sleep Hours']].values
y = data['Result']

w = np.array([0.5,0.3])
b= -2
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def neuron(X,w,b):
  z = np.dot(X, w) + b
  return sigmoid(z)
y_pred_prob = neuron(X,w,b)
```

```python
y_pred = (y_pred_prob > 0.5).astype(int)
```

```python
accuracy = np.mean(y_pred == y)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
plt.scatter(range(len(y)),y,label="Actual",alpha=0.6)
plt.scatter(range(len(y)),y_pred,label="Predicted",alpha=0.6)
plt.legend()
plt.title("Actual vs Predicted")
plt.show()
```
